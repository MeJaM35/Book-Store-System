```python
import torch
import torch.nn as nn
from accelerate import Accelerator
from transformers import AutoTokenizer, AutoConfig, AutoModelForCausalLM

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-2-7b-chat-hf",device_map="auto", load_in_8bit=True,cache_dir= "C:/Users/TUF/OneDrive/Desktop/Github/Model_1")
```

```python
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-2-7b-chat-hf", cache_dir= "C:/Users/TUF/OneDrive/Desktop/Github/Model_1")
```

```python
input = tokenizer("She is a", return_tensors="pt").to(device)
input
```

```python
output = model.generate(**input,max_new_tokens=10)
```

```python
response = tokenizer.decode(output[0])
```
